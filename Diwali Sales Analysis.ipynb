```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sadee\\Downloads\\Diwali Sales Data.csv",encoding = 'unicode_escape')
df.head(10)
```

# Data Cleaning

```python
df.drop(['Status','unnamed1'],inplace=True,axis=1)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

# Exploratory Data Analysis (EDA)

```python
sns.countplot(x = 'Gender',data = df)
```

Sales With Respect To Gender

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(data=sales_gen,x='Gender',y='Amount')
```

Sales With Respect To Age Group

```python
countplot = sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in countplot.containers:
    countplot.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y='Amount')
```

Sales With Respect To State

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sales_state
```

```python
barplot = sns.barplot(y='State',x='Orders',data=sales_state)
```

```python
df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

Occupation

```python
countplot = sns.countplot(y='Occupation',data=df)
for bars in countplot.containers:
    countplot.bar_label(bars)
```

Product Category

```python
countplot = sns.countplot(y='Product_Category',data=df,palette='Greens')
for bars in countplot.containers:
    countplot.bar_label(bars)
```

```python
sales_by_category = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(y='Product_Category',x='Amount',data=sales_by_category)
```
